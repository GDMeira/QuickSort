```python
import numpy as np
import random
import math
```

Troca

```python
def swap(x,i,j):
  aux = x[i]
  x[i] = x[j]
  x[j] = aux
  return
```

Partição

```python
def partition(x,l,u,p):
  global troca
  global comparacao
  infer = l-1
  sup = u+1
  find_inf = False
  find_sup = False
  while infer < p or sup > p:
    comparacao += 2
    while not find_inf and infer < p:
      infer += 1
      comparacao += 1
      if x[infer]>x[p]:
        comparacao += 1
        find_inf = True
    while not find_sup and sup > p:
      sup -= 1
      comparacao += 1
      if x[sup]<x[p]:
        find_sup = True
    if find_inf and find_sup:
      swap(x,infer,sup)
      troca += 1
      find_inf = False
      find_sup = False
    elif find_inf:
      swap(x,infer,p)
      troca += 1
      p = infer
      find_inf = False
    elif find_sup:
      swap(x,sup,p)
      troca += 1
      p = sup
      find_sup = False
  return p
```

QuickSort

```python
def QuikSort(x,l,u,p):
  global troca
  global comparacao
  if l >= u:
    comparacao +=1
    return
  p = partition(x,l,u,p)
  p_inf = math.floor((l+p-1)/2)
  QuikSort(x,l,p-1,p_inf)
  p_sup = math.floor((p+1+u)/2)
  QuikSort(x,p+1,u,p_sup)
  return
N = int(input('Entre o tamanho do vetor aleatorio: '))
x = random.sample(range(0,50),N)
x_antes = list(x)

troca = 0
comparacao = 0


QuikSort(x,0,N-1,math.floor(N/2))

print('O nro de comparacoes foi: ',comparacao)
print('O nro de trocas foi: ',troca)
print('O vetor inicial era: ')
for i in range(len(x_antes)):
  print(f"{x_antes[i]} ", end ='')
print('\nO vetor ordenado eh: ')
for i in range (len(x)):
  print(f"{x[i]} ",end='')
```

```python
def partition_lista(x,l,u,p):
  global troca
  global comparacao
  infer = l-1
  sup = u+1
  find_inf = False
  find_sup = False
  while infer < p or sup > p:
    comparacao += 2
    while not find_inf and infer < p:
      infer += 1
      comparacao += 1
      if x[infer]>x[p]:
        comparacao += 1
        find_inf = True
    while not find_sup and sup > p:
      sup -= 1
      comparacao += 1
      if x[sup]<x[p]:
        find_sup = True
    if find_inf and find_sup:
      x.troca(infer,sup)
      troca += 1
      find_inf = False
      find_sup = False
    elif find_inf:
      x.troca(infer,p)
      troca += 1
      p = infer
      find_inf = False
    elif find_sup:
      x.troca(p,sup)
      troca += 1
      p = sup
      find_sup = False
  return p

def QuikSort_lista(x,l,u,p):
  global troca
  global comparacao
  if l >= u:
    comparacao +=1
    return
  p = partition_lista(x,l,u,p)
  p_inf = math.floor((l+p-1)/2)
  QuikSort_lista(x,l,p-1,p_inf)
  p_sup = math.floor((p+1+u)/2)
  QuikSort_lista(x,p+1,u,p_sup)
  return

class No:
  def __init__(self,dado):
    self.dado = dado
    self.prox = None
    self.ant = None

class ListaDupla:
  def __init__(self):
    self.inicio = None
    self.fim = None
    self._tamanho = 0

  def add_inicio(self,elem):
    no = No(elem)
    if self.inicio: #lista nao vazia
      no.prox = self.inicio #novo no aponta pro antigo 1 no como prox
      self.inicio.ant = no #antigo 1 no aponta pro novo no como anterior
      self.inicio = no
    else: #lista esta vazia
      self.inicio = self.fim = no
    self._tamanho += 1

  def add_final(self,elem):
    no = No(elem)
    if self.fim: #lista nao vazia
      no.ant = self.fim #novo no aponta pro antigo final como anterior
      self.fim.prox = no #antigo no final aponta pro novo no como proximo
      self.fim = no
    else: #lista vazia
      self.inicio = self.fim = no
    self._tamanho += 1

  def troca(self,i,j): #trocar por sequencia de remocao e adicao
      if i-1>self._tamanho or j-1>self._tamanho:
        raise IndexError("Indice maior que tamanho da lista")
      if i>j: #ir do menor pro maior
        aux = i
        i = j
        j = aux
      p = self.inicio
      for k in range(j): #leva o ponteiro ate o no j
        if k == i:
          pi = p
        p = p.prox
      pj = p
      if j == i+1 and i == 0:
        pj.prox.ant = pi
        pj.ant = pi.ant
        pi.prox = pj.prox
        pj.prox = pi
        pi.ant = pj
        self.inicio = pj
      elif j == i+1 and j == self._tamanho-1:
        pi.ant.prox = pj
        pj.ant = pi.ant
        pi.prox = pj.prox
        pj.prox = pi
        pi.ant = pj
        self.fim = pi
      elif j == i+1:
        pi.ant.prox = pj
        pj.prox.ant = pi
        pj.ant = pi.ant
        pi.prox = pj.prox
        pj.prox = pi
        pi.ant = pj
      elif i>0 and j<self._tamanho-1:
        pi.ant.prox = pj
        pi.prox.ant = pj
        pj.ant.prox = pi
        pj.prox.ant = pi
        aux1 = pj.ant
        aux2 = pj.prox
        pj.ant = pi.ant
        pj.prox = pi.prox
        pi.ant = aux1
        pi.prox = aux2
      elif i==0 and j<self._tamanho-1:
        pi.prox.ant = pj
        pj.ant.prox = pi
        pj.prox.ant = pi
        aux1 = pj.ant
        aux2 = pj.prox
        pj.ant = pi.ant
        pj.prox = pi.prox
        pi.ant = aux1
        pi.prox = aux2
        self.inicio = pj
      elif i>0 and j==self._tamanho-1:
        pi.ant.prox = pj
        pi.prox.ant = pj
        pj.ant.prox = pi
        aux1 = pj.ant
        aux2 = pj.prox
        pj.ant = pi.ant
        pj.prox = pi.prox
        pi.ant = aux1
        pi.prox = aux2
        self.fim = pi
      else:
        pi.prox.ant = pj
        pj.ant.prox = pi
        aux1 = pj.ant
        aux2 = pj.prox
        pj.ant = pi.ant
        pj.prox = pi.prox
        pi.ant = aux1
        pi.prox = aux2
        self.fim = pi
        self.inicio = pj

  def __len__(self):
      return self._tamanho

  def __getitem__(self,index):
      if index > self._tamanho:
        raise IndexError("Indice maior que o tamanho da lista")
      p = self.inicio
      for i in range(index):
        p = p.prox
      return p.dado

  def __setitem__(self,index,elem):
      if index > self._tamanho:
        raise IndexError("Indice maior que o tamanho da lista")
      p = self.inicio
      for i in range(index):
        p = p.prox
      p.dado = elem

  def mostrar(self):
      p = self.inicio
      for i in range(self._tamanho):
        print(f"{p.dado} ",end='')
        p = p.prox

troca = 0
comparacao = 0

N = int(input('Entre o tamanho da lista: '))
lista = ListaDupla()
for i in range(N):
  lista.add_inicio(random.randint(0,50))
print("\nA lista original eh: ")
lista.mostrar()
QuikSort_lista(lista,0,N-1,math.floor(N/2))
#lista.troca(0,3)
print("\nA lista organizada eh:")
lista.mostrar()
print('\nO nro de comparacoes foi: ',comparacao)
print('\nO nro de trocas foi: ',troca)
```

Ordenação de registros

```python
def partition_registro(x,l,u,p):
  global troca
  global comparacao
  infer = l-1
  sup = u+1
  find_inf = False
  find_sup = False
  while infer < p or sup > p:
    comparacao += 2
    while not find_inf and infer < p:
      infer += 1
      comparacao += 1
      if x[infer]>x[p]:
        comparacao += 1
        find_inf = True
    while not find_sup and sup > p:
      sup -= 1
      comparacao += 1
      if x[sup]<x[p]:
        find_sup = True
    if find_inf and find_sup:
      x.troca(infer,sup)
      troca += 1
      find_inf = False
      find_sup = False
    elif find_inf:
      x.troca(infer,p)
      troca += 1
      p = infer
      find_inf = False
    elif find_sup:
      x.troca(p,sup)
      troca += 1
      p = sup
      find_sup = False
  return p

def QuikSort_registro(x,l,u,p):
  global troca
  global comparacao
  if l >= u:
    comparacao +=1
    return
  p = partition_registro(x,l,u,p)
  p_inf = math.floor((l+p-1)/2)
  QuikSort_registro(x,l,p-1,p_inf)
  p_sup = math.floor((p+1+u)/2)
  QuikSort_registro(x,p+1,u,p_sup)
  return

class Registro:
  def __init__(self,dado,char1,char2,real1,real2):
    self.dado = dado
    self.char1 = char1
    self.char2 = char2
    self.real1 = real1
    self.real2 = real2

class VetorRegistros:
  def __init__(self,N):
    self._tamanho = N
    self.v = [None]*N

  def add_final(self,elem,char1,char2,real1,real2):
    reg = Registro(elem,char1,char2,real1,real2)
    for i in range(self._tamanho):
      if self.v[i] == None:
        self.v[i] = reg
        return
    raise Exception("O vetor de registros ja esta cheio.")

  def troca(self,i,j): #trocar por sequencia de remocao e adicao
    aux1 = self.v[i]
    self.v[i] = self.v[j]
    self.v[j] = aux1

  def __len__(self):
    return self._tamanho

  def __getitem__(self,index):
    if index > self._tamanho:
      raise IndexError("Indice maior que o tamanho da lista")
    return self.v[index].dado

  def mostrar(self):
    for i in range(self._tamanho):
        print(self.v[i].dado,"  ",self.v[i].char1,"  ",self.v[i].char2,"  ",self.v[i].real1,"  ",self.v[i].real2,"  ")

troca = 0
comparacao = 0

a = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','x','y','z']
N = int(input('Entre o tamanho da lista: '))
r = VetorRegistros(N)
for i in range(N):
  r.add_final(random.randint(0,50),''.join(random.choices(a,k=50)),''.join(random.choices(a,k=50)),random.random()*N,random.random()*N)
print("Os registros originais sao: ")
r.mostrar()
QuikSort_registro(r,0,N-1,math.floor(N/2))
#lista.troca(0,3)
print("\nOs registros organizados sao:")
r.mostrar()
print('\nO nro de comparacoes foi: ',comparacao)
print('\nO nro de trocas foi: ',troca)
```
